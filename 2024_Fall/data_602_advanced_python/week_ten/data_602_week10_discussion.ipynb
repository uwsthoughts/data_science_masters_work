## 1. What is matplottlib and seaborn?  When would you choose one over the other?

Answer:
Matplotlib is a library for creating visualizations, including:
 * Pie charts,
 * Bar charts
 * Line charts 
 * Histograms
 
These visualizations are highest customizable and can be static or interactive.

Seaborn is built on Matplotlib and specializes in statistical data visualizations, including:
* Heatmaps
* Box plots 
* Violin plots
* Pair plots 

Matplotlib is better for when you need highly customized visual elements since it comes with a lot more user-control. Seaborn is the better choice when you are dealing with statistical data since it allows for easy creation of quickly and visually pleasing visualizations with less code than Matplotlib. Typically, I'd choose Matplotlib for complex plotting needs and Seaborn for streamlined, statistical visuals.

## 3. Give an example of either a matplotlib or seaborn graphic (give code).  You may also reference an informative article.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
tips = sns.load_dataset("tips")

# Create a simple Seaborn box plot that focuses on essential data elements
plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", data=tips)

# Add title and labels, and remove unnecessary spines to maintain high data-ink ratio
plt.title("Bill distribution by day of week.")
plt.ylabel("Dollars")
sns.despine()  # Removes top and right borders for a cleaner look

plt.show()
```
