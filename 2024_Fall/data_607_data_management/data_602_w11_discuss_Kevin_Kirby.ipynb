```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## 1. How do you plot a histogram in Seaborn? 

Answer: You can plot a histogram in Seaborn using the `sns.histplot()` function.  Here is an example:


```python
citi_df = pd.read_parquet('https://storage.googleapis.com/data_science_masters_files/2024_fall/data_602_advanced_python/week_elevent/data_606_citiBike.parquet')
sns.histplot(data=citi_df, x='rideable_type')
```

```python
citi_df.head()
```

## 2. Plot a histogram with NAs dropped.

Answer: 
NAs are dropped using `dropna()`. Here's an example:

```python
sns.histplot(data=citi_df.dropna(subset=['member_casual']), x='member_casual')
```

## 3. How do you set the color for a histogram?

Answer: 
The color is set using the `color` parameter in `sns.histplot()`. Here's an example:

```python
sns.histplot(data=citi_df.dropna(subset=['member_casual']), x='member_casual', color='red')
```

## 4.  What type of plot would allow you to compare two continuous features?  Give an example of code.

Answer:
You can use a scatter plot to compare two continuous features. Here's an example with the sarting latitudes and longitudes of Citi Bike trips:

```python
sns.scatterplot(data=citi_df, x='start_lat', y='start_lng')
```

## 5. Give example of a correlation plot.

Answer:
A correlation plot is a scatter plot where the data is jittered and the transparency is adjusted using the `alpha` parameter. Here's an example:

```python
sns.scatterplot(data=citi_df, x='start_lat', y='start_lng', alpha=0.1)
```

## 6. Change the figure size of your plot(s).

Answer:
The figure size is set using the `figsize` parameter in `plt.figure()`. Here's an example:


```python
plt.figure(figsize=(10, 5))
sns.scatterplot(data=citi_df, x='start_lat', y='start_lng', alpha=0.1)
```
