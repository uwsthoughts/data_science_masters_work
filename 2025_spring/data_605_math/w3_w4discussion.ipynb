## Week 3 Discussion

I chose exercise C23 from page 306 of A First Course in Linear Algebra:

Find the eigenvalues, eigenspaces, algebraic and geometric multiplicities for 

A =

[1 1]

[1 1]

```python
import numpy as np

A = np.array([[1, 1], [1, 1]])

e_values, e_vectors = np.linalg.eig(A)

alg_m = [np.sum(e_values == eig) for eig in e_values]
geo_m = [np.linalg.matrix_rank(e_vectors[:, e_values == eig]) for eig in e_values]

e_spaces = [e_vectors[:, i] for i in range(len(e_values))]

for i, e in enumerate(e_values):
    print(f"Eigenvalue: {e}")
    print(f"Eigenvector: {e_spaces[i]}")
    print(f"Algebraic multiplicity: {alg_m[i]}")
    print(f"Geometric multiplicity: {geo_m[i]}")
```

## Week 4 Discussion

I chose exercise T10 on page 469 in chapter R of the first course linear algebra book.

"Suppose that T:V→VT:V→V is an invertible linear transformation with a nonzero eigenvalue λ. Prove that 1/λ is an eigenvalue of T^-1"


```python
import numpy as np
import matplotlib.pyplot as plt

lambda_2 = 3
V = np.array([1, 1])  
print(f"V: {V}")

T = np.array([[lambda_2, 0], [0, lambda_2]])      
print(f"T: {T}")

inverse_t = np.linalg.inv(T)      
print(f"inverse_t: {inverse_t}")

T_V = T @ V
print(f"T_V: {T_V}")

inv_tv = inverse_t @ V
print(f"inv_tv: {inv_tv}")

fig, ax = plt.subplots()
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)

ax.quiver(0, 0, V[0], V[1], angles='xy', scale_units='xy', scale=1, color='blue', label="Starting V")
ax.quiver(0, 0, T_V[0], T_V[1], angles='xy', scale_units='xy', scale=1, color='pink', label="T(V) = λV", alpha=0.7)
ax.quiver(0, 0, inv_tv[0], inv_tv[1], angles='xy', scale_units='xy', scale=1, color='skyblue', label="T⁻¹(V) = (1/λ)V")

ax.legend(loc='lower right')
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)
ax.set_title("Kevin's Proof")

plt.grid()
plt.show()
```
