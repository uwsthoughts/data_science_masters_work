```python
import pandas as tian
# tian.set_option("display.max_rows", None)
tian.set_option('display.max_columns', None)
import seaborn as sns
import builtins
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import squarify
import textwrap

```

```python
def df_deets(df_or_names):
    def print_info(df, df_name):
        print(f"{df_name}:")
        if hasattr(builtins, "display"):
            display(df)
        else:
            print(df)
        print("\n")
        
        summary_df = tian.DataFrame({
            "Data types": df.dtypes,
            "Duplicates": [df.duplicated(subset=[col]).sum() for col in df.columns],
            "NAs": df.isna().sum().values,
            "Distinct values": df.nunique().values
        }).set_index(df.columns)
        
        print(f"{df_name} summary:")
        if hasattr(builtins, "display"):
            display(summary_df)
        else:
            print(summary_df)
        print("\n")
        print(df.columns)
        print("\n")
        total_duplicates = df.duplicated().sum()
        print(f"Total number of duplicate rows in {df_name}: {total_duplicates}\n")

    if isinstance(df_or_names, list): 
        for item in df_or_names:
            if isinstance(item, str):
                df = globals()[item] 
                print_info(df, item)
            else:
                print_info(item, "dataframe")
    else: 
        if isinstance(df_or_names, str):
            df = globals()[df_or_names]  
            print_info(df, df_or_names)
        else:
            print_info(df_or_names, "dataframe")
```

```python
bp_track_df = tian.read_csv('/Users/uwsthoughts/Desktop/music_project_files/csv_data/bp_track.csv')
bp_genre_df = tian.read_csv('/Users/uwsthoughts/Desktop/music_project_files/csv_data/bp_genre.csv')
```

```python
df_deets('bp_track_df')
```

```python
df_deets('bp_genre_df')
```

```python
bp_track_df["release_date"] = tian.to_datetime(bp_track_df["release_date"])
bp_track_df.dtypes
```

```python
bp_track_df["release_date"].dt.year.unique()
```

```python
yearly_songs = bp_track_df.groupby(bp_track_df["release_date"].dt.year).agg(
    first_release=("release_date", "min"),
    last_release=("release_date", "max"),
    total_tracks=("track_id", "count"),
    distinct_isrcs=("isrc", "nunique")
).reset_index().rename(columns={"release_date": "year"}).sort_values("year", ascending=False)
yearly_songs
```

```python
recent_years = yearly_songs[(yearly_songs["year"] >= 2013) & (yearly_songs["year"] <= 2022)]
recent_years
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=recent_years, x="year", y="distinct_isrcs")

plt.xlabel("Year")
plt.ylabel("Total Songs")
plt.title("Total Songs Released on Beatport - 2013-2022")

plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, pos: f"{int(x):,}"))

plt.show()
```

```python
genre_counts = bp_genre_df.groupby('genre_name')['song_count'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.scatterplot(data=genre_counts, x='genre_name', y='song_count', size='song_count', sizes=(20, 1000), alpha=0.6, legend=False)

plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))  # Format y-axis

plt.xlabel("Genre")
plt.ylabel("Song Count")
plt.title("Song Count by Genre (Bubble Size Proportional to Count)")
plt.show()
```

```python

plt.figure(figsize=(12, 8))
squarify.plot(sizes=genre_counts['song_count'], label=genre_counts['genre_name'], alpha=0.7)
plt.axis('off')
plt.title("Song Count by Genre (Treemap)")
plt.show()
```

```python

# Wrap long genre names to fit within their respective blocks
genre_counts['genre_name'] = genre_counts['genre_name'].apply(lambda x: '\n'.join(textwrap.wrap(x, width=10)))

plt.figure(figsize=(12, 6))
sns.scatterplot(data=genre_counts, x='genre_name', y='song_count', size='song_count', sizes=(20, 1000), alpha=0.6, legend=False)

plt.xticks(rotation=0, ha='center')  # No rotation, keeps labels centered
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"{int(x):,}"))
plt.xlabel("Genre")
plt.ylabel("Song Count")
plt.title("Song Count by Genre (Bubble Size Proportional to Count)")
plt.show()
```

```python

genre_counts_bars = bp_genre_df.groupby('genre_name')['song_count'].sum().reset_index().sort_values('song_count', ascending=False)

plt.figure(figsize=(10, 12))
sns.barplot(data=genre_counts_bars, y='genre_name', x='song_count', palette='Blues_r')

plt.xlabel("Song Count")
plt.ylabel("Genre")
plt.title("Song Count by Genre")
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"{int(x):,}"))  # Format numbers
plt.show()
```

Preconscious vision refers to the ability of the human brain to process visual information rapidly and effortlessly, often before we are consciously aware of it. In the context of data visualization, leveraging preconscious vision can help convey complex information quickly and intuitively. Effective visual designs capitalize on this by using shapes, patterns, and color contrasts that our brains can recognize instantly, making the data more accessible and easier to interpret.
