```python
from dotenv import load_dotenv
load_dotenv("/Users/kevinkirby/Desktop/local_files/variables/.env")

import nltk
from nltk.corpus import gutenberg
from collections import Counter
import matplotlib.pyplot as plt
```

```python
gutenberg_words = [word.lower() for word in gutenberg.words('melville-moby_dick.txt') if word.isalpha()]

guten_counts = Counter(gutenberg_words)
guten_totals = sum(guten_counts.values())
guten_sorted = guten_counts.most_common()
guten_unique = len(guten_counts)

print(f"Total unique words: {guten_unique}")
```

```python
guten_200 = guten_sorted[:200]

running_total = 0
for idx, (_, count) in enumerate(guten_sorted, 1):
    running_total += count
    if running_total >= guten_totals / 2:
        unique_halfs = idx
        break

print(f"Unique words representing half the corpus: {unique_halfs}")
```

```python
guten_200 = guten_sorted[:200]

print(guten_200)
```

```python
words, counts = zip(*guten_200)
rel_freqs = [count / guten_totals for count in counts]

plt.figure(figsize=(10, 22))
plt.barh(words[::-1], rel_freqs[::-1])
plt.ylabel('Word', labelpad=40)
plt.xlabel('Relative Frequency')
plt.title('Relative Frequencies of Top 200 Words')
plt.tight_layout()
plt.show()
```

Does the observed relative frequency of these words follow Zipfs law? Explain.

Answer:

The relative frequency plot definitely looks like what Zipfs law predicts. The most common words are way more frequent than the rest. After the top few, the frequencies drop off super fast and keep flattening out. That steep curve at the start is a classic Zipfs law pattern.
