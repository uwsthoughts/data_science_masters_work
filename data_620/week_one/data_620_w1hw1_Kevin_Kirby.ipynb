```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
G.add_edges_from([
 ('A','B'), ('B','C'), ('C','D'), ('D','A'),
 ('F','G'), ('G','I'), ('I','H'), ('H','F'),
 ('C','E'), ('E','H')
])
nx.draw(G, with_labels=True)
plt.show()
```
