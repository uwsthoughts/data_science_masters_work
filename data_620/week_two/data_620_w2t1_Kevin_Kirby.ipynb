```python
from torch_geometric.datasets import Planetoid
from torch_geometric.utils import to_networkx
import networkx as nx
import matplotlib.pyplot as plt
```

```python
planettoid = Planetoid(root='/tmp/Cora', name='Cora')
planettoid_data = planettoid[0]
network_graph = to_networkx(planettoid_data, to_undirected=True)
largest_cc = max(nx.connected_components(network_graph), key=len)
network_graph_sub = network_graph.subgraph(largest_cc).copy()

diameter = nx.diameter(network_graph_sub)
clustering_coeff = nx.average_clustering(network_graph_sub)

print(f"Diameter: {diameter}")
print(f"Clustering Coefficient: {clustering_coeff}")
```

```python
planettoid = Planetoid(root='/tmp/Cora', name='Cora')
planettoid_data = planettoid[0]
network_graph = to_networkx(planettoid_data, to_undirected=True)
largest_cc = max(nx.connected_components(network_graph), key=len)
network_graph_sub = network_graph.subgraph(largest_cc).copy()

plt.figure(figsize=(10, 10))
pos = nx.spring_layout(network_graph_sub, seed=42)
nx.draw(network_graph_sub, pos, node_size=20, edge_color='gray', alpha=0.7)
plt.title("Cora Largest Connected Component")
plt.axis("off")
plt.show()
```
